```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('dark_background')

df = pd.read_csv('netflix_titles.csv')
df = df.drop(df.columns[0], axis=1)
```

```python
mask = df['type'] == 'Movie'
```

```python
df2 = pd.DataFrame(df[mask])

eixo_x = df2['listed_in'].value_counts().keys()
eixo_y = df2['listed_in'].value_counts().values
```

```python
plt.bar(eixo_x, eixo_y)

plt.title('Filmes por gênero')

plt.xticks(rotation=90)

plt.xlabel('Gêneros de filmes')
plt.ylabel('Quantidade de filmes')

plt.show()
```

```python
eixo_x = df2['release_year'].value_counts().sort_index().keys()
eixo_y = df2['release_year'].value_counts().sort_index().values
```

```python
plt.plot(eixo_x, eixo_y)

plt.title('Filmes por ano de Lançamento')
plt.xlabel('Ano de lançamento')
plt.ylabel('Quantidade de filmes')

plt.show()
```
